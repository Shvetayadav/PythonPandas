```python
df1 = sns.load_dataset('tips')
```

```python
df1[['total_bill', 'tip']].iplot(x = 'total_bill', y= 'tip', mode = 'markers', size = 6)
```

```python
df1.iplot(kind = 'scatter3d', x = 'total_bill', y= 'tip', z = 'size' )
```

```python
df1.iplot(kind = 'bubble3d', x = 'total_bill', y= 'tip', z = 'size', size = 'tip' )
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data  = df1)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data  = df1, hue  = 'smoker')
```

```python
sns.relplot(x = 'total_bill', y='tip', data =df1, style = 'size')
```
